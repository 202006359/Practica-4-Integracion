```python
import socket
import pickle
import threading
import time
import functions as mf

MCAST_GRP = '224.1.1.1'
MCAST_PORT = 5007
mf.clear_current_connections()
def main():
    # Elija el numero de jugadores en la partida antes de iniciar el juego
    num_jugadores = 2

    #Elija el numero de preguntas que se realizaran en la partida
    num_preguntas = 5
    
    multicast_sock = mf.join_multicast_group()
#     print("Servidor iniciado.")

    #Esperar a que se conecten los jugadores  
    mf.wait_for_players(multicast_sock, num_jugadores)

    # Tras 1 segundo de espera, se inicia el juego
    time.sleep(1)
    mf.init_game(multicast_sock, num_jugadores, num_preguntas)
    multicast_sock.close()
    
def handle_client(connection, client_address):
    error_message=''
    try:
        while True:  # Mantener la conexión abierta
            data = connection.recv(1024).decode()
            if not data:
                break  # Romper el bucle si no se recibe data
            else:
                print (data)
                command, username, password = data.split(' ')
            if command == 'C':
#                 fs.remove_connection(username, client_address)
                break  # Romper el bucle si el comando es 'exit'
            elif command == 'B':
                if mf.register_user(username, password):
                    connection.sendall('Registro exitoso\n'.encode())
                else:
                    error_message='Error: el usuario ya existe'
                    break
            elif command == 'A':
                if mf.login_user(username, password):
                    mf.add_connection(username, str(client_address))
                    connection.sendall('Login exitoso\n'.encode())
                    main()
                    break
                else:
                    error_message='Error: usuario o contraseña incorrectos'
                    break
            else:
                break
    finally:
        connection.sendall(('Desconectado. Se le ha desconectado del servidor'+ error_message).encode())
        connection.close()
        

# Configuración del servidor
server_address = ('localhost', 6783)
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
sock.bind(server_address)
sock.listen()

print("Servidor iniciado, esperando conexiones...")
while True:
    connection, client_address = sock.accept()
    print(f"Conexión desde {client_address}")
    threading.Thread(target=handle_client, args=(connection, client_address)).start()



# if __name__ == "__main__":
#     main()
```
